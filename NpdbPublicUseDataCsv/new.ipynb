# Data Preparation: Dtype Correction & Validation

**Reference:** PublicUseDataFile-Format.pdf (December 31, 2025 edition), Pages 9-10

This cell corrects all dtype mismatches between the official NPDB format specification and pandas' auto-inferred types.

### Key Mismatches Found:

| Issue | Variables | Problem | Fix |
|-------|-----------|---------|-----|
| **Dollar → object** | `PAYMENT`, `TOTALPMT` | Loaded as string (`'$3500'`) instead of numeric | Strip `$`/`,` → `float64` |
| **Numeric codes → float64** | `PRACTAGE`, `GRAD`, `ALGNNATR`, `ALEGATN1`, `ALEGATN2`, `OUTCOME`, `MALYEAR1`, `MALYEAR2`, `NUMBPRSN`, `PTAGE`, `AAYEAR`, `AACLASS1-5`, `AALENGTH`, `AAEFYEAR`, `AASIGYR`, `FUNDPYMT` | NaN forces float; should be nullable integer | Convert to `Int64` (pandas nullable integer) |
| **Deprecated fields** | `WORKCTRY`, `HOMECTRY` | Deprecated since June 2017; loaded as `float64` | Convert to `string`, fill NaN |
| **String → category** | `RECTYPE`, `WORKSTAT`, `HOMESTAT`, `LICNSTAT`, `PAYNUMBR`, `PAYTYPE`, `PYRRLTNS`, `PTSEX`, `PTTYPE`, `AALENTYP`, `BASISCD1-5` | Low-cardinality strings waste memory | Convert to `category` |
| **RECTYPE values** | `RECTYPE` | Official values are M/P/A/C (not M/N as some sources say) | Verified from data: M=old mal, **P**=new mal, A=old AA, C=new AA |


````python
import pandas as pd
import numpy as np

# ============================================================
# STEP 0: Load the raw data
# ============================================================
df = pd.read_csv(
    r"C:\Users\Owner\Desktop\MTSU_2026_Spring\NPDB Project\NpdbPublicUseDataCsv\NPDB2510.CSV",
    low_memory=False
)
print(f"Loaded: {df.shape[0]:,} rows x {df.shape[1]} columns")
print(f"\nRECTYPE value counts:")
print(df['RECTYPE'].value_counts().to_string())

# ============================================================
# STEP 1: Fix PAYMENT and TOTALPMT (Dollar strings → numeric)
# ============================================================
# These are stored as '$3500', '$195000', etc. in the CSV
for col in ['PAYMENT', 'TOTALPMT']:
    if df[col].dtype == object:
        df[col] = (
            df[col]
            .astype(str)
            .str.replace('$', '', regex=False)
            .str.replace(',', '', regex=False)
            .replace('nan', np.nan)
        )
        df[col] = pd.to_numeric(df[col], errors='coerce')
        print(f"  Fixed {col}: object → {df[col].dtype}")

# ============================================================
# STEP 2: Fix Numeric codes that have NaN (float64 → Int64)
# ============================================================
# These should be integers but NaN forces float64.
# pandas Int64 (capital I) supports nullable integers.
nullable_int_cols = [
    'PRACTAGE', 'GRAD', 'ALGNNATR', 'ALEGATN1', 'ALEGATN2',
    'OUTCOME', 'MALYEAR1', 'MALYEAR2', 'NUMBPRSN', 'PTAGE',
    'AAYEAR', 'AACLASS1', 'AACLASS2', 'AACLASS3', 'AACLASS4', 'AACLASS5',
    'AAEFYEAR', 'AASIGYR', 'FUNDPYMT'
]
for col in nullable_int_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce').round().astype('Int64')
print(f"\n  Converted {len(nullable_int_cols)} columns to Int64 (nullable integer)")

# AALENGTH stays float64 (can have decimal values like 0.5 years)
if 'AALENGTH' in df.columns:
    df['AALENGTH'] = pd.to_numeric(df['AALENGTH'], errors='coerce')
    print(f"  AALENGTH kept as {df['AALENGTH'].dtype} (decimal years)")

# ============================================================
# STEP 3: Fix deprecated String fields (WORKCTRY, HOMECTRY)
# ============================================================
# These were deprecated since June 2017 per the PDF.
# They were loaded as float64 because they contain mostly NaN.
for col in ['WORKCTRY', 'HOMECTRY']:
    if col in df.columns:
        df[col] = df[col].astype(str).replace('nan', pd.NA).astype('string')
        print(f"  Fixed {col}: float64 → string (deprecated field)")

# ============================================================
# STEP 4: Convert low-cardinality string columns to category
# ============================================================
category_cols = [
    'RECTYPE', 'WORKSTAT', 'HOMESTAT', 'LICNSTAT',
    'PAYNUMBR', 'PAYTYPE', 'PYRRLTNS', 'PTSEX', 'PTTYPE',
    'AALENTYP', 'BASISCD1', 'BASISCD2', 'BASISCD3', 'BASISCD4', 'BASISCD5'
]
for col in category_cols:
    if col in df.columns:
        df[col] = df[col].astype('category')
cat_memory = sum(df[col].memory_usage(deep=True) for col in category_cols if col in df.columns)
print(f"\n  Converted {len(category_cols)} columns to category dtype")

# ============================================================
# STEP 5: Verify & Display Final dtypes
# ============================================================
print("\n" + "="*70)
print("FINAL DTYPE SUMMARY (after corrections)")
print("="*70)

# Build comparison table
official_types = {
    'SEQNO': 'Numeric', 'RECTYPE': 'String', 'REPTYPE': 'Numeric', 'ORIGYEAR': 'Numeric',
    'WORKSTAT': 'String', 'WORKCTRY': 'String', 'HOMESTAT': 'String', 'HOMECTRY': 'String',
    'LICNSTAT': 'String', 'LICNFELD': 'Numeric', 'PRACTAGE': 'Numeric', 'GRAD': 'Numeric',
    'ALGNNATR': 'Numeric', 'ALEGATN1': 'Numeric', 'ALEGATN2': 'Numeric', 'OUTCOME': 'Numeric',
    'MALYEAR1': 'Numeric', 'MALYEAR2': 'Numeric', 'PAYMENT': 'Dollar', 'TOTALPMT': 'Dollar',
    'PAYNUMBR': 'String', 'NUMBPRSN': 'Numeric', 'PAYTYPE': 'String', 'PYRRLTNS': 'String',
    'PTAGE': 'Numeric', 'PTSEX': 'String', 'PTTYPE': 'String', 'AAYEAR': 'Numeric',
    'AACLASS1': 'Numeric', 'AACLASS2': 'Numeric', 'AACLASS3': 'Numeric',
    'AACLASS4': 'Numeric', 'AACLASS5': 'Numeric',
    'BASISCD1': 'String', 'BASISCD2': 'String', 'BASISCD3': 'String',
    'BASISCD4': 'String', 'BASISCD5': 'String',
    'AALENTYP': 'String', 'AALENGTH': 'Numeric', 'AAEFYEAR': 'Numeric', 'AASIGYR': 'Numeric',
    'TYPE': 'Numeric', 'PRACTNUM': 'Numeric', 'ACCRRPTS': 'Numeric',
    'NPMALRPT': 'Numeric', 'NPLICRPT': 'Numeric', 'NPCLPRPT': 'Numeric',
    'NPPSMRPT': 'Numeric', 'NPDEARPT': 'Numeric', 'NPEXCRPT': 'Numeric',
    'NPGARPT': 'Numeric', 'NPCTMRPT': 'Numeric', 'FUNDPYMT': 'Numeric'
}

print(f"{'Column':<12} {'Official':>10} {'Pandas Dtype':>16} {'Status':>8}")
print("-" * 48)
for col in df.columns:
    off = official_types.get(col, '?')
    pdt = str(df[col].dtype)
    # Check if the mapping is correct
    ok = True
    if off in ('Numeric', 'Dollar') and pdt not in ('int64', 'Int64', 'float64'):
        ok = False
    if off == 'String' and pdt not in ('category', 'object', 'string'):
        ok = False
    status = 'OK' if ok else 'ISSUE'
    print(f"{col:<12} {off:>10} {pdt:>16} {status:>8}")

# ============================================================
# STEP 6: Memory usage comparison
# ============================================================
mem_mb = df.memory_usage(deep=True).sum() / 1e6
print(f"\nTotal memory usage: {mem_mb:.1f} MB")
print(f"Records: {len(df):,}")

# ============================================================
# STEP 7: Data quality summary
# ============================================================
print("\n" + "="*70)
print("DATA QUALITY SUMMARY")
print("="*70)

# RECTYPE breakdown
print("\nRecord Type Distribution:")
for val, label in [('M','Malpractice (old format, pre-2004)'),
                    ('P','Malpractice (new format, post-2004)'),
                    ('A','Adverse Action (old format, pre-1999)'),
                    ('C','Adverse Action (CAAR, post-1999)')]:
    ct = (df['RECTYPE'] == val).sum()
    print(f"  {val}: {ct:>10,} ({ct/len(df)*100:5.1f}%) — {label}")

malp = df[df['RECTYPE'].isin(['M', 'P'])]
print(f"\nMalpractice Payment Records (M+P): {len(malp):,}")
print(f"  PAYMENT non-null: {malp['PAYMENT'].notna().sum():,}")
print(f"  PAYMENT range: ${malp['PAYMENT'].min():,.0f} — ${malp['PAYMENT'].max():,.0f}")
print(f"  PAYMENT mean: ${malp['PAYMENT'].mean():,.0f}")
print(f"  PAYMENT median: ${malp['PAYMENT'].median():,.0f}")

# Missing value summary for key malpractice columns
print("\nMissing Values in Malpractice Records (M+P):")
mal_cols = ['PAYMENT','TOTALPMT','ALGNNATR','ALEGATN1','OUTCOME',
            'MALYEAR1','PAYTYPE','PRACTAGE','PTAGE','PTSEX','WORKSTAT']
for col in mal_cols:
    if col in malp.columns:
        n_miss = malp[col].isna().sum()
        pct = n_miss / len(malp) * 100
        print(f"  {col:<12}: {n_miss:>8,} missing ({pct:5.1f}%)")

print("\n✅ Data preparation complete. All dtypes corrected per official format.")

````

## Understanding NaN Values: Malpractice vs. Adverse Action Reports

The NPDB dataset is **not a single homogeneous table**. Each row is one of **two fundamentally different kinds of reports**, identified by the `RECTYPE` column:

| RECTYPE | Report Category | What It Captures |
|---------|----------------|------------------|
| **M** or **P** | **Malpractice Payment Report** | A financial payment was made on behalf of a healthcare practitioner to settle or satisfy a malpractice claim (lawsuit, settlement, or judgment). |
| **A** or **C** | **Adverse Action Report** | A disciplinary/administrative action was taken against a practitioner — license revocation, suspension, clinical privilege restriction, DEA action, Medicare exclusion, etc. |

These are completely different events with completely different attributes, but they are **stored in the same flat CSV file** — meaning every row has all 54 columns, even though roughly **half the columns only apply to one type**.

---

### Why This Creates NaN Values

#### When `RECTYPE` = M or N (Malpractice Payment)
The report describes a **financial payout** from a malpractice claim, so these columns are **populated**:
- `ALGNNATR`, `ALEGATN1`, `ALEGATN2` — what was alleged (e.g., misdiagnosis, surgical error)
- `OUTCOME` — how severely the patient was injured
- `MALYEAR1`, `MALYEAR2` — when the malpractice occurred
- `PAYMENT`, `TOTALPMT`, `PAYNUMBR` — payment details
- `PAYTYPE`, `PYRRLTNS`, `FUNDPYMT` — who paid and how
- `PTAGE`, `PTSEX`, `PTTYPE` — patient demographics

But the **adverse action columns are NaN** because no disciplinary action is being reported:
- `AAYEAR`, `AACLASS1–5`, `BASISCD1–5`, `AALENTYP`, `AALENGTH`, `AAEFYEAR`, `AASIGYR` → all **NaN**

#### When `RECTYPE` = A or C (Adverse Action)
The report describes a **disciplinary action**, so these columns are **populated**:
- `AAYEAR` — when the action was taken
- `AACLASS1–5` — what type of action (e.g., revocation, suspension, probation)
- `BASISCD1–5` — why (e.g., incompetence, substance abuse, criminal conviction)
- `AALENTYP`, `AALENGTH` — how long the penalty lasts
- `AAEFYEAR`, `AASIGYR` — when it took effect

But the **malpractice-specific columns are NaN** because no payment event is being reported:
- `ALGNNATR`, `ALEGATN1`, `ALEGATN2`, `OUTCOME`, `MALYEAR1`, `MALYEAR2` → all **NaN**
- `PAYMENT`, `TOTALPMT`, `PAYNUMBR`, `PAYTYPE`, `PYRRLTNS`, `FUNDPYMT` → all **NaN**
- `PTAGE`, `PTSEX`, `PTTYPE` → all **NaN**

### Key Implications for Analysis

1. **Don't treat NaNs as "missing data" to impute** — they are *structurally absent*. A malpractice payment row *should not* have an `AACLASS1` value.
2. **Filter by `RECTYPE` before analyzing** — e.g., `df[df['RECTYPE'].isin(['M','P'])]` for malpractice-only analysis, or `df[df['RECTYPE'].isin(['A','C'])]` for adverse actions.
3. **This explains the high NaN counts** — columns like `ALGNNATR` (1,365,463 NaNs) and `AAYEAR` (529,830 NaNs) aren't poorly collected data; they reflect the ~72%/28% split between the two report types in the dataset.

## 1. Introduction & Motivation

Medical malpractice occurs when a healthcare professional's treatment falls below the accepted **standard of care**, resulting in injury or death to a patient. The **National Practitioner Data Bank (NPDB)**, established by Congress under Title IV of Public Law 99-660 (the Health Care Quality Improvement Act of 1986), serves as a national repository that tracks malpractice payments and adverse actions against healthcare practitioners across the United States.

This report traces the **lifecycle of a medical malpractice lawsuit** — from the initial incident to final compensation — and maps each stage to the variables captured in the NPDB Public Use Data File (NPDB2510.CSV, containing **1,895,122 records** spanning **1990–2025**).

> **Key Reference:** U.S. Department of Health and Human Services, HRSA. *National Practitioner Data Bank Public Use Data File Documentation.* Available at: [https://www.npdb.hrsa.gov/resources/publicData.jsp](https://www.npdb.hrsa.gov/resources/publicData.jsp)

---

## 2. The Lifecycle of a Medical Malpractice Lawsuit

The malpractice process follows a well-defined sequence. Each stage is described below with its corresponding NPDB variables.

### Stage 1: The Medical Incident (Act or Omission)

**What Happens:**
A healthcare practitioner commits an act or omission that deviates from the accepted standard of care. This could be a misdiagnosis, surgical error, medication mistake, failure to monitor, or any other clinical failure that results in patient harm.

**NPDB Variables Capturing This Stage:**

| Variable | What It Records |
|----------|----------------|
| `MALYEAR1` | The **year the act/omission first occurred** — marks the beginning of the malpractice event |
| `MALYEAR2` | The **year the act/omission last occurred** — captures events that span multiple years (e.g., ongoing mismanagement) |
| `ALGNNATR` | The **allegation group** — broad category of what went wrong (e.g., Diagnosis-Related, Surgery-Related, Medication-Related) |
| `ALEGATN1` | The **primary specific allegation** — detailed code for the exact nature of the malpractice |
| `ALEGATN2` | A **secondary allegation**, if applicable (e.g., both a surgical error AND a failure to obtain informed consent) |

**Key Insight from the Data:** The gap between `MALYEAR1` and `ORIGYEAR` (year reported to NPDB) reveals the **lag between the incident and resolution** — often spanning many years due to discovery, litigation, and appeals.

---

### Stage 2: Patient Injury / Outcome

**What Happens:**
The patient suffers harm as a direct result of the practitioner's negligence. The severity can range from minor emotional distress to permanent disability or death. **Establishing causation** (proving the practitioner's action directly caused the injury) is one of the four legal elements required for a malpractice claim.

**The Four Elements of Medical Malpractice (Legal Standard):**
1. **Duty** — A doctor-patient relationship existed
2. **Breach** — The practitioner violated the standard of care
3. **Causation** — The breach directly caused the injury
4. **Damages** — The patient suffered quantifiable harm

**NPDB Variables Capturing This Stage:**

| Variable | What It Records |
|----------|----------------|
| `OUTCOME` | The **severity of injury** to the patient, coded on a scale (e.g., Emotional Injury Only → Minor Temporary → Major Temporary → Minor Permanent → Major Permanent → Quadriplegic/Brain Damage → Death) |
| `PTAGE` | **Patient's age** at the time of the event |
| `PTSEX` | **Patient's sex** (Male, Female, Unknown) |
| `PTTYPE` | **Patient type** — whether the patient was an Inpatient, Outpatient, or other category |
| `NUMBPRSN` | Number of practitioners involved on whose behalf a payment was made |

**Key Insight:** The `OUTCOME` variable is critical for understanding the relationship between injury severity and payment amounts. More severe outcomes (death, brain damage) typically correspond to larger payments.

---

### Stage 3: Filing the Claim

**What Happens:**
The injured patient (or their representative) files a formal written complaint or claim demanding monetary compensation. Most states require filing within a **statute of limitations** (typically 1–3 years from the date of injury or discovery of injury). Many states also require a **Certificate of Merit** — a sworn statement from a qualified medical expert confirming that malpractice likely occurred — before a lawsuit can proceed.

**Pre-Suit Requirements (vary by state):**
- Medical record review by an expert
- Certificate of merit / affidavit of merit
- Pre-suit notice to the healthcare provider
- Some states require pre-suit mediation panels

**NPDB Variables Capturing This Stage:**

| Variable | What It Records |
|----------|----------------|
| `LICNSTAT` | The **state of licensure** — determines which state's laws govern the claim (statutes of limitations, damage caps, certificate of merit requirements) |
| `WORKSTAT` | The **state where the practitioner was working** — often the jurisdiction where the incident occurred and where the lawsuit is filed |
| `LICNFELD` | The **practitioner's field of license** — identifies the medical specialty involved |
| `PRACTAGE` | **Practitioner's age** at the time of the incident |
| `GRAD` | **Practitioner's graduation year** — can be used to estimate years of experience |

**Key Insight:** State-level variation is enormous. Some states have strict damage caps (e.g., California's MICRA limited non-economic damages to $250,000 until 2023), while others have none. The `WORKSTAT` and `LICNSTAT` variables allow state-by-state analysis of these differences.

> **Reference:** Bal, B.S. (2009). "An Introduction to Medical Malpractice in the United States." *Clinical Orthopaedics and Related Research*, 467(2), 339-347.

---

### Stage 4: Discovery, Litigation & Expert Testimony

**What Happens:**
This is the longest phase, often lasting **1–3 years or more**. Both sides exchange evidence through:
- **Interrogatories** — written questions answered under oath
- **Depositions** — recorded oral testimony from parties, witnesses, and expert witnesses
- **Document production** — medical records, hospital policies, training records, correspondence
- **Expert witness review** — both plaintiff and defense retain medical experts to opine on standard of care

**Why NPDB doesn't capture this stage directly:**
The NPDB records **outcomes** (payments and adverse actions), not the litigation process itself. However, we can infer litigation timelines:

| Inference | How to Calculate |
|-----------|-----------------|
| **Time from incident to report** | `ORIGYEAR - MALYEAR1` — approximates how long the legal process took |
| **Multi-payment cases** | `PAYNUMBR` — sequential payment numbers indicate prolonged or multi-stage resolutions |
| **Multiple practitioners** | `NUMBPRSN` — cases involving multiple defendants may indicate more complex litigation |

---

### Stage 5: Resolution — Settlement vs. Judgment

**What Happens:**
A malpractice claim resolves in one of three ways:
1. **Settlement** — the most common outcome (~93-97% of malpractice payment cases settle out of court). Both parties agree on a payment amount without going to trial.
2. **Judgment/Verdict** — the case goes to trial, and a judge or jury awards damages. Plaintiffs win approximately 20-30% of cases that go to trial.
3. **Arbitration** — a neutral third party makes a binding decision.

**NPDB Variables Capturing This Stage:**

| Variable | What It Records |
|----------|----------------|
| `PAYTYPE` | **How the payment was resolved**: Settlement, Judgment/Verdict, or Other |
| `PYRRLTNS` | **Who paid**: The payer's relationship to the practitioner (e.g., insurance company, self-insured entity, state fund) |
| `FUNDPYMT` | Whether a **state patient compensation fund** contributed to or made the payment |

**Key Insight:** The `PAYTYPE` variable reveals the settlement-vs-trial ratio. The fact that most cases settle explains why many payment amounts cluster below the median — parties often settle for less than the potential trial award to avoid uncertainty and legal costs.

> **Reference:** Jena, A.B., et al. (2011). "Outcomes of Medical Malpractice Litigation Against US Physicians." *Archives of Internal Medicine*, 171(18), 1006-1011.

---

### Stage 6: Compensation / Payment

**What Happens:**
Once a case resolves via settlement or judgment, payment is made on behalf of the practitioner to the injured patient. Under federal law (42 U.S.C. § 11131), **any malpractice payment made on behalf of a named practitioner must be reported to the NPDB within 30 days**, regardless of amount. This is what generates the NPDB record.

**Types of Damages Awarded:**
- **Economic Damages** — medical expenses, lost wages, rehabilitation costs, future care needs
- **Non-Economic Damages** — pain and suffering, emotional distress, loss of consortium, loss of enjoyment of life
- **Punitive Damages** — rare; awarded only for egregious misconduct to punish the defendant

**NPDB Variables Capturing This Stage:**

| Variable | What It Records |
|----------|----------------|
| `PAYMENT` | The **specific payment amount** made on behalf of the practitioner in this report |
| `TOTALPMT` | The **total cumulative payment** for the entire claim/event |
| `PAYNUMBR` | The **sequential number** of this payment (e.g., 1 of 3 payments for the same event) |
| `ORIGYEAR` | The **year the report was processed** into the NPDB — effectively the year payment was made (since reports are due within 30 days of payment) |
| `RECTYPE` | **M** (old malpractice format, pre-2004) or **N** (new format, post-2004) |
| `REPTYPE` | Specific report type code (e.g., 301 = Medical Malpractice Payment Report) |

**Key Insight:** The `PAYMENT` vs. `TOTALPMT` distinction matters. A single malpractice event may involve multiple payments (e.g., initial payment + supplemental payment), each generating its own report. `TOTALPMT` captures the full picture while `PAYMENT` captures the individual report's amount.

> **Reference:** National Practitioner Data Bank. "Reporting Medical Malpractice Payments." HRSA. Available at: [https://www.npdb.hrsa.gov/guidebook/EReporting.jsp](https://www.npdb.hrsa.gov/guidebook/EReporting.jsp)

---

### Stage 7: Post-Payment — Practitioner Record & Consequences

**What Happens:**
After a payment is reported, it becomes part of the practitioner's **permanent NPDB record**. Hospitals, licensing boards, and other eligible entities query the NPDB when making credentialing, privileging, and licensing decisions. A practitioner with multiple malpractice reports may face:
- License restrictions or revocations (triggers an Adverse Action report)
- Loss of hospital privileges
- Increased malpractice insurance premiums
- Exclusion from Medicare/Medicaid

**NPDB Variables Capturing This Stage:**

| Variable | What It Records |
|----------|----------------|
| `PRACTNUM` | **De-identified practitioner ID** — allows tracking the same practitioner across multiple reports |
| `TYPE` | **Practitioner type** (e.g., MD/DO, Dentist, Nurse, etc.) |
| `ACCRRPTS` | **Total accumulated reports** in the NPDB for this practitioner |
| `NPMALRPT` | Number of **Malpractice Payment Reports** on file |
| `NPLICRPT` | Number of **Licensure Action Reports** on file |
| `NPCLPRPT` | Number of **Clinical Privileges Reports** on file |
| `NPPSMRPT` | Number of **Professional Society Reports** on file |
| `NPDEARPT` | Number of **DEA Action Reports** on file |
| `NPEXCRPT` | Number of **Medicare/Medicaid Exclusion Reports** on file |
| `NPGARPT` | Number of **Government Administrative Reports** on file |
| `NPCTMRPT` | Number of **Criminal Conviction Reports** on file |

**Key Insight:** The `ACCRRPTS` and individual report count variables allow analysis of **repeat offenders** — practitioners with multiple malpractice payments. Research shows a small percentage of physicians account for a disproportionate share of malpractice payments.

> **Reference:** Studdert, D.M., et al. (2016). "Prevalence and Characteristics of Physicians Prone to Malpractice Claims." *New England Journal of Medicine*, 374(4), 354-362.

## 5. Important Caveats & Limitations

| Caveat | Explanation |
|--------|-------------|
| **Only paid claims are reported** | If a malpractice lawsuit is dismissed or the defendant wins at trial with no payment, it does NOT appear in the NPDB. This creates a **selection bias** — the data only reflects cases that resulted in payment. |
| **No litigation details** | The NPDB does not capture the discovery process, trial proceedings, or the specific legal arguments. It only captures the final outcome. |
| **Payment ≠ Guilt** | A malpractice payment does not imply the practitioner was negligent. Many cases settle to avoid litigation costs regardless of merit. |
| **De-identified data** | Public Use files do not contain practitioner names, hospital names, or specific dates — only years and coded categories. |
| **Underreporting risk** | While federal law mandates reporting, compliance is not perfect, particularly for smaller payments or in certain states. |
| **State law variation** | Damage caps, pre-suit requirements, and statutes of limitations vary dramatically by state, affecting all payment distributions. |

# Module 3: Temporal Trends (Inflation-Adjusted)

Analyzes how malpractice payments have evolved over time, adjusting for inflation using CPI data. This addresses the question: are payments actually growing, or is it just inflation?


````python
# --- Ensure PAYMENT/TOTALPMT are numeric (auto-fix if data prep cell was skipped) ---
import numpy as np
for _col in ['PAYMENT', 'TOTALPMT']:
    if _col in df.columns and df[_col].dtype == object:
        df[_col] = pd.to_numeric(
            df[_col].astype(str).str.replace('$','',regex=False).str.replace(',','',regex=False).replace('nan',np.nan),
            errors='coerce'
        )

# --- CPI Adjustment Factors (Bureau of Labor Statistics, CPI-U, base year 2025) ---
# Source: BLS CPI-U annual averages (approximate multipliers to 2025 dollars)
cpi_factors = {
    1990: 2.32, 1991: 2.22, 1992: 2.16, 1993: 2.10, 1994: 2.04, 1995: 1.99,
    1996: 1.93, 1997: 1.89, 1998: 1.86, 1999: 1.82, 2000: 1.76, 2001: 1.71,
    2002: 1.69, 2003: 1.65, 2004: 1.60, 2005: 1.55, 2006: 1.50, 2007: 1.46,
    2008: 1.41, 2009: 1.42, 2010: 1.39, 2011: 1.35, 2012: 1.32, 2013: 1.30,
    2014: 1.28, 2015: 1.28, 2016: 1.26, 2017: 1.24, 2018: 1.21, 2019: 1.18,
    2020: 1.17, 2021: 1.12, 2022: 1.03, 2023: 1.00, 2024: 0.97, 2025: 0.95
}

# Apply CPI adjustment
mal['CPI_FACTOR'] = mal['ORIGYEAR'].map(cpi_factors)
mal['PAYMENT_ADJ'] = mal['PAYMENT'] * mal['CPI_FACTOR']

# --- Annual Trends ---
annual = mal.groupby('ORIGYEAR').agg(
    count=('PAYMENT', 'size'),
    total_nominal=('PAYMENT', 'sum'),
    total_adj=('PAYMENT_ADJ', 'sum'),
    mean_nominal=('PAYMENT', 'mean'),
    mean_adj=('PAYMENT_ADJ', 'mean'),
    median_nominal=('PAYMENT', 'median'),
    median_adj=('PAYMENT_ADJ', 'median')
).reset_index()

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Claim count by year
axes[0,0].bar(annual['ORIGYEAR'], annual['count'], color='steelblue', alpha=0.8)
axes[0,0].set_title('Number of Malpractice Payment Reports by Year', fontsize=13, fontweight='bold')
axes[0,0].set_xlabel('Year')
axes[0,0].set_ylabel('Count')

# 2. Total payments: nominal vs adjusted
axes[0,1].plot(annual['ORIGYEAR'], annual['total_nominal']/1e9, label='Nominal $', color='gray', linewidth=1.5)
axes[0,1].plot(annual['ORIGYEAR'], annual['total_adj']/1e9, label='2025 Dollars', color='steelblue', linewidth=2)
axes[0,1].set_title('Total Annual Payments (Billions)', fontsize=13, fontweight='bold')
axes[0,1].set_xlabel('Year')
axes[0,1].set_ylabel('$ Billions')
axes[0,1].legend()

# 3. Mean payment: nominal vs adjusted
axes[1,0].plot(annual['ORIGYEAR'], annual['mean_nominal'], label='Nominal $', color='gray', linewidth=1.5)
axes[1,0].plot(annual['ORIGYEAR'], annual['mean_adj'], label='2025 Dollars', color='darkorange', linewidth=2)
axes[1,0].set_title('Mean Payment per Claim', fontsize=13, fontweight='bold')
axes[1,0].set_xlabel('Year')
axes[1,0].set_ylabel('$ Amount')
axes[1,0].legend()

# 4. Median payment: nominal vs adjusted
axes[1,1].plot(annual['ORIGYEAR'], annual['median_nominal'], label='Nominal $', color='gray', linewidth=1.5)
axes[1,1].plot(annual['ORIGYEAR'], annual['median_adj'], label='2025 Dollars', color='green', linewidth=2)
axes[1,1].set_title('Median Payment per Claim', fontsize=13, fontweight='bold')
axes[1,1].set_xlabel('Year')
axes[1,1].set_ylabel('$ Amount')
axes[1,1].legend()

plt.suptitle('Malpractice Payment Temporal Trends (1990-2025)', fontsize=15, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()

# --- Year-over-Year Changes ---
annual['pct_chg_count'] = annual['count'].pct_change() * 100
annual['pct_chg_mean_adj'] = annual['mean_adj'].pct_change() * 100
print("\nYear-over-Year Changes (selected columns):")
print(annual[['ORIGYEAR','count','mean_adj','median_adj','pct_chg_count','pct_chg_mean_adj']].tail(15).to_string(index=False))

# --- 5-Year Rolling Averages ---
annual['rolling_mean_adj'] = annual['mean_adj'].rolling(5, center=True).mean()
annual['rolling_median_adj'] = annual['median_adj'].rolling(5, center=True).mean()

fig, ax = plt.subplots(figsize=(14, 5))
ax.plot(annual['ORIGYEAR'], annual['mean_adj'], alpha=0.4, color='darkorange', label='Annual Mean (2025$)')
ax.plot(annual['ORIGYEAR'], annual['rolling_mean_adj'], color='darkorange', linewidth=2.5, label='5-Year Rolling Mean')
ax.plot(annual['ORIGYEAR'], annual['median_adj'], alpha=0.4, color='green', label='Annual Median (2025$)')
ax.plot(annual['ORIGYEAR'], annual['rolling_median_adj'], color='green', linewidth=2.5, label='5-Year Rolling Median')
ax.set_title('Rolling Average Payment Trends (2025 Dollars)', fontsize=13, fontweight='bold')
ax.set_xlabel('Year')
ax.set_ylabel('Payment ($)')
ax.legend()
plt.tight_layout()
plt.show()
````

# Module 8: Settlement vs. Judgment Analysis

Compares payments by resolution type (`PAYTYPE`). Research shows ~93-97% of paid malpractice cases settle; we verify this and test whether judgments yield different payment amounts.


````python
# --- Ensure PAYMENT/TOTALPMT are numeric (auto-fix if data prep cell was skipped) ---
import numpy as np
for _col in ['PAYMENT', 'TOTALPMT']:
    if _col in df.columns and df[_col].dtype == object:
        df[_col] = pd.to_numeric(
            df[_col].astype(str).str.replace('$','',regex=False).str.replace(',','',regex=False).replace('nan',np.nan),
            errors='coerce'
        )

# PAYTYPE mapping
paytype_map = {
    '1': 'Judgment/Verdict', 'A': 'Judgment/Verdict',
    '2': 'Settlement', 'B': 'Settlement',
    '3': 'Arbitration', 'C': 'Arbitration',
    '4': 'Other/Not Classified', 'D': 'Other/Not Classified'
}

pay_data = mal.dropna(subset=['PAYTYPE']).copy()
pay_data['PAYTYPE_STR'] = pay_data['PAYTYPE'].astype(str).str.strip()
pay_data['resolution'] = pay_data['PAYTYPE_STR'].map(paytype_map).fillna('Unknown')

# --- Frequency Table ---
res_stats = pay_data.groupby('resolution').agg(
    count=('PAYMENT', 'size'),
    mean=('PAYMENT', 'mean'),
    median=('PAYMENT', 'median'),
    total=('PAYMENT', 'sum')
).reset_index()
res_stats['pct'] = (res_stats['count'] / res_stats['count'].sum() * 100).round(1)
res_stats = res_stats.sort_values('count', ascending=False)

print("Payment by Resolution Type:")
print(f"{'Type':<24} {'Count':>10} {'Pct':>6} {'Mean ($)':>14} {'Median ($)':>14} {'Total ($M)':>12}")
print("-" * 82)
for _, row in res_stats.iterrows():
    print(f"{row['resolution']:<24} {row['count']:>10,} {row['pct']:>5.1f}% {row['mean']:>14,.0f} {row['median']:>14,.0f} {row['total']/1e6:>12,.1f}")

# --- Visualization ---
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Pie chart of resolution types
axes[0].pie(res_stats['count'], labels=res_stats['resolution'], autopct='%1.1f%%',
            colors=['#2196F3', '#FF9800', '#4CAF50', '#9E9E9E'], startangle=90)
axes[0].set_title('Resolution Type Distribution', fontsize=13, fontweight='bold')

# Mean payment comparison
colors_bar = ['#2196F3', '#FF9800', '#4CAF50', '#9E9E9E'][:len(res_stats)]
axes[1].bar(res_stats['resolution'], res_stats['mean'], color=colors_bar, alpha=0.85)
axes[1].set_title('Mean Payment by Resolution Type', fontsize=13, fontweight='bold')
axes[1].set_ylabel('Mean Payment ($)')
plt.setp(axes[1].xaxis.get_majorticklabels(), rotation=20, ha='right')

# Box plot
sns.boxplot(data=pay_data, x='resolution', y='PAYMENT', showfliers=False, palette='Set2', ax=axes[2])
axes[2].set_title('Payment Distribution by Resolution (outliers hidden)', fontsize=13, fontweight='bold')
axes[2].set_ylabel('Payment ($)')
axes[2].set_ylim(0, mal['PAYMENT'].quantile(0.95))
plt.setp(axes[2].xaxis.get_majorticklabels(), rotation=20, ha='right')

plt.tight_layout()
plt.show()

# --- Statistical Test: Settlement vs Judgment ---
settlements = pay_data[pay_data['resolution'] == 'Settlement']['PAYMENT']
judgments = pay_data[pay_data['resolution'] == 'Judgment/Verdict']['PAYMENT']
u_stat, p_val = stats.mannwhitneyu(settlements, judgments, alternative='two-sided')
print(f"\nMann-Whitney U Test (Settlement vs. Judgment):")
print(f"  U-statistic: {u_stat:,.0f}")
print(f"  p-value:     {p_val:.2e}")
print(f"  Conclusion:  {'Significantly different' if p_val < 0.05 else 'Not significantly different'} at alpha=0.05")
print(f"  Settlement median: ${settlements.median():,.0f} | Judgment median: ${judgments.median():,.0f}")
````

# Module 9: Catastrophic Claims Analysis (>$1M)

Following Bixenstine et al. (2014), we define catastrophic claims as those with payments exceeding $1 million. The paper found catastrophic claims = 7.9% of all claims but 36.2% of total dollars (2004-2010). We update these figures through 2025.


````python
# --- Ensure PAYMENT/TOTALPMT are numeric (auto-fix if data prep cell was skipped) ---
import numpy as np
for _col in ['PAYMENT', 'TOTALPMT']:
    if _col in df.columns and df[_col].dtype == object:
        df[_col] = pd.to_numeric(
            df[_col].astype(str).str.replace('$','',regex=False).str.replace(',','',regex=False).replace('nan',np.nan),
            errors='coerce'
        )

# --- Define Catastrophic Claims ---
CATASTROPHIC_THRESHOLD = 1_000_000
mal['is_catastrophic'] = mal['PAYMENT'] >= CATASTROPHIC_THRESHOLD

n_total = len(mal)
n_cat = mal['is_catastrophic'].sum()
total_dollars = mal['PAYMENT'].sum()
cat_dollars = mal[mal['is_catastrophic']]['PAYMENT'].sum()

print("="*70)
print("CATASTROPHIC CLAIMS ANALYSIS (Payment >= $1,000,000)")
print("="*70)
print(f"Total malpractice claims:     {n_total:>10,}")
print(f"Catastrophic claims:          {n_cat:>10,} ({n_cat/n_total*100:.1f}%)")
print(f"Total dollars paid:           ${total_dollars:>16,.0f}")
print(f"Catastrophic dollars:         ${cat_dollars:>16,.0f} ({cat_dollars/total_dollars*100:.1f}%)")
print(f"Max single payment:           ${mal['PAYMENT'].max():>16,.0f}")
print(f"Mean catastrophic payment:    ${mal[mal['is_catastrophic']]['PAYMENT'].mean():>16,.0f}")
print(f"Median catastrophic payment:  ${mal[mal['is_catastrophic']]['PAYMENT'].median():>16,.0f}")

# --- Bixenstine et al. Comparison ---
bix_period = mal[(mal['ORIGYEAR'] >= 2004) & (mal['ORIGYEAR'] <= 2010)]
bix_cat = bix_period[bix_period['is_catastrophic']]
print(f"\n--- Bixenstine Replication (2004-2010) ---")
print(f"Total claims: {len(bix_period):,}")
print(f"Catastrophic: {len(bix_cat):,} ({len(bix_cat)/len(bix_period)*100:.1f}%) [Paper: 7.9%]")
print(f"Dollar share: {bix_cat['PAYMENT'].sum()/bix_period['PAYMENT'].sum()*100:.1f}% [Paper: 36.2%]")

# --- Catastrophic Claims Over Time ---
cat_annual = mal.groupby('ORIGYEAR').agg(
    total_claims=('is_catastrophic', 'size'),
    cat_claims=('is_catastrophic', 'sum'),
    total_dollars=('PAYMENT', 'sum')
).reset_index()
cat_annual['cat_pct'] = cat_annual['cat_claims'] / cat_annual['total_claims'] * 100
cat_annual['cat_dollars'] = mal[mal['is_catastrophic']].groupby('ORIGYEAR')['PAYMENT'].sum().reindex(cat_annual['ORIGYEAR']).fillna(0).values
cat_annual['cat_dollar_pct'] = cat_annual['cat_dollars'] / cat_annual['total_dollars'] * 100

fig, axes = plt.subplots(2, 2, figsize=(16, 10))

axes[0,0].bar(cat_annual['ORIGYEAR'], cat_annual['cat_claims'], color='crimson', alpha=0.8)
axes[0,0].set_title('Number of Catastrophic Claims by Year', fontsize=12, fontweight='bold')
axes[0,0].set_ylabel('Count')

axes[0,1].plot(cat_annual['ORIGYEAR'], cat_annual['cat_pct'], color='crimson', linewidth=2, marker='o', markersize=3)
axes[0,1].set_title('Catastrophic Claims as % of All Claims', fontsize=12, fontweight='bold')
axes[0,1].set_ylabel('Percentage (%)')
axes[0,1].axhline(y=cat_annual['cat_pct'].mean(), color='gray', linestyle='--', alpha=0.5, label=f"Mean: {cat_annual['cat_pct'].mean():.1f}%")
axes[0,1].legend()

axes[1,0].bar(cat_annual['ORIGYEAR'], cat_annual['cat_dollars']/1e9, color='darkred', alpha=0.8)
axes[1,0].set_title('Total Catastrophic Dollars by Year', fontsize=12, fontweight='bold')
axes[1,0].set_ylabel('$ Billions')

axes[1,1].plot(cat_annual['ORIGYEAR'], cat_annual['cat_dollar_pct'], color='darkred', linewidth=2, marker='o', markersize=3)
axes[1,1].set_title('Catastrophic $ as % of Total $', fontsize=12, fontweight='bold')
axes[1,1].set_ylabel('Percentage (%)')
axes[1,1].axhline(y=cat_annual['cat_dollar_pct'].mean(), color='gray', linestyle='--', alpha=0.5, label=f"Mean: {cat_annual['cat_dollar_pct'].mean():.1f}%")
axes[1,1].legend()

plt.suptitle('Catastrophic Claims (>$1M) Trends', fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout()
plt.show()

# --- Profile of Catastrophic Claims ---
cat_claims = mal[mal['is_catastrophic']].copy()
print("\n--- Top States for Catastrophic Claims ---")
top_cat_states = cat_claims.groupby('WORKSTAT').size().sort_values(ascending=False).head(10)
for state, count in top_cat_states.items():
    print(f"  {state}: {count:,}")

print("\n--- Top Allegation Groups for Catastrophic Claims ---")
cat_allg = cat_claims.dropna(subset=['ALGNNATR'])
cat_allg['ALGNNATR'] = pd.to_numeric(cat_allg['ALGNNATR'], errors='coerce').astype('Int64')
cat_allg['allg_label'] = cat_allg['ALGNNATR'].map(allg_map).fillna('Unknown')
top_cat_allg = cat_allg.groupby('allg_label').size().sort_values(ascending=False).head(8)
for allg, count in top_cat_allg.items():
    print(f"  {allg}: {count:,}")

print("\n--- Outcome Distribution for Catastrophic Claims ---")
cat_out = cat_claims.dropna(subset=['OUTCOME'])
cat_out['outcome_label'] = cat_out['OUTCOME'].map({
    1: 'Emotional', 2: 'Insignificant', 3: 'Minor Temp', 4: 'Major Temp',
    5: 'Minor Perm', 6: 'Significant Perm', 7: 'Major Perm',
    8: 'Quad/Brain', 9: 'Death', 10: 'Cannot Determine'
}).fillna('Unknown')
for outcome, count in cat_out.groupby('outcome_label').size().sort_values(ascending=False).items():
    pct = count / len(cat_out) * 100
    print(f"  {outcome}: {count:,} ({pct:.1f}%)")
````

# Module 10: Time-to-Payment Duration Analysis

Analyzes the lag between the medical incident (`MALYEAR1`) and the payment report year (`ORIGYEAR`). The Frees & Gao paper notes this lag is typically 5-7 years and creates difficulties in attributing effects to policy changes.


````python
# --- Ensure PAYMENT/TOTALPMT are numeric (auto-fix if data prep cell was skipped) ---
import numpy as np
for _col in ['PAYMENT', 'TOTALPMT']:
    if _col in df.columns and df[_col].dtype == object:
        df[_col] = pd.to_numeric(
            df[_col].astype(str).str.replace('$','',regex=False).str.replace(',','',regex=False).replace('nan',np.nan),
            errors='coerce'
        )

# --- Compute Lag ---
lag_data = mal.dropna(subset=['MALYEAR1']).copy()
lag_data['lag_years'] = lag_data['ORIGYEAR'] - lag_data['MALYEAR1']
# Remove unreasonable values
lag_data = lag_data[(lag_data['lag_years'] >= 0) & (lag_data['lag_years'] <= 50)]

print("="*70)
print("TIME FROM INCIDENT TO PAYMENT (years)")
print("="*70)
print(f"Records with valid lag: {len(lag_data):,}")
print(f"Mean lag:   {lag_data['lag_years'].mean():.1f} years")
print(f"Median lag: {lag_data['lag_years'].median():.1f} years")
print(f"Mode lag:   {lag_data['lag_years'].mode().iloc[0]:.0f} years")
print(f"Std Dev:    {lag_data['lag_years'].std():.1f} years")
print(f"Max lag:    {lag_data['lag_years'].max():.0f} years")

# Percentile table
for p in [25, 50, 75, 90, 95, 99]:
    print(f"  {p}th percentile: {lag_data['lag_years'].quantile(p/100):.0f} years")

# --- Lag Distribution ---
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

axes[0].hist(lag_data['lag_years'], bins=range(0, 35), color='steelblue', edgecolor='white', alpha=0.85)
axes[0].axvline(lag_data['lag_years'].mean(), color='red', linestyle='--', label=f"Mean: {lag_data['lag_years'].mean():.1f} yrs")
axes[0].axvline(lag_data['lag_years'].median(), color='orange', linestyle='--', label=f"Median: {lag_data['lag_years'].median():.1f} yrs")
axes[0].set_title('Distribution of Incident-to-Payment Lag', fontsize=13, fontweight='bold')
axes[0].set_xlabel('Years from Incident to Payment')
axes[0].set_ylabel('Frequency')
axes[0].legend()

# Lag trend over time (has it gotten shorter?)
lag_by_year = lag_data.groupby('ORIGYEAR')['lag_years'].agg(['mean','median']).reset_index()
axes[1].plot(lag_by_year['ORIGYEAR'], lag_by_year['mean'], label='Mean Lag', color='steelblue', linewidth=2)
axes[1].plot(lag_by_year['ORIGYEAR'], lag_by_year['median'], label='Median Lag', color='darkorange', linewidth=2)
axes[1].set_title('Lag Trend Over Time: Is Resolution Getting Faster?', fontsize=13, fontweight='bold')
axes[1].set_xlabel('Payment Year')
axes[1].set_ylabel('Lag (years)')
axes[1].legend()

plt.tight_layout()
plt.show()

# --- Lag by Injury Severity ---
lag_sev = lag_data.dropna(subset=['OUTCOME']).copy()
lag_sev['outcome_label'] = lag_sev['OUTCOME'].map({
    1: '1-Emotional', 2: '2-Insignificant', 3: '3-Minor Temp', 4: '4-Major Temp',
    5: '5-Minor Perm', 6: '6-Significant Perm', 7: '7-Major Perm',
    8: '8-Quad/Brain', 9: '9-Death', 10: '10-Cannot Det.'
}).fillna('Unknown')

lag_sev_stats = lag_sev.groupby('outcome_label')['lag_years'].agg(['mean','median','count']).reset_index()
lag_sev_stats = lag_sev_stats.sort_values('outcome_label')
print("\nLag by Injury Severity:")
print(f"{'Outcome':<24} {'Count':>8} {'Mean Lag':>10} {'Median Lag':>12}")
print("-" * 56)
for _, row in lag_sev_stats.iterrows():
    if 'Cannot' not in row['outcome_label'] and 'Unknown' not in row['outcome_label']:
        print(f"{row['outcome_label']:<24} {row['count']:>8,} {row['mean']:>10.1f} {row['median']:>12.0f}")

# --- Lag by State (Top 10) ---
lag_state = lag_data.groupby('WORKSTAT')['lag_years'].agg(['mean','median','count']).reset_index()
lag_state = lag_state[lag_state['count'] >= 100].sort_values('mean', ascending=False)
print("\nTop 10 States by Mean Lag (min 100 claims):")
for _, row in lag_state.head(10).iterrows():
    print(f"  {row['WORKSTAT']}: mean={row['mean']:.1f} yrs, median={row['median']:.0f} yrs (n={row['count']:,})")

# --- Lag vs Payment Amount ---
fig, ax = plt.subplots(figsize=(12, 5))
lag_payment = lag_data.groupby('lag_years').agg(
    mean_payment=('PAYMENT', 'mean'),
    median_payment=('PAYMENT', 'median'),
    count=('PAYMENT', 'size')
).reset_index()
lag_payment = lag_payment[lag_payment['count'] >= 20]  # min sample

ax.bar(lag_payment['lag_years'], lag_payment['median_payment'], color='steelblue', alpha=0.5, label='Median')
ax.plot(lag_payment['lag_years'], lag_payment['mean_payment'], color='darkorange', linewidth=2, marker='o', markersize=4, label='Mean')
ax.set_title('Payment Amount by Lag Duration', fontsize=13, fontweight='bold')
ax.set_xlabel('Lag (Years from Incident to Payment)')
ax.set_ylabel('Payment ($)')
ax.legend()
plt.tight_layout()
plt.show()

print("\n--- KEY INSIGHT ---")
print(f"The Frees & Gao paper cited 5-7 year typical lag. Our full dataset shows:")
print(f"  Mean: {lag_data['lag_years'].mean():.1f} years | Median: {lag_data['lag_years'].median():.1f} years")
print(f"  This {'confirms' if 4.5 <= lag_data['lag_years'].mean() <= 8 else 'differs from'} the paper's estimate.")
````
